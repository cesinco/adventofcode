```python
def move_tail(head_node, tail_node):
    dist_vec = [head_node[0] - tail_node[0], head_node[1] - tail_node[1]]
    if abs(dist_vec[0]) > 1 and abs(dist_vec[1]) > 1:
        if head_node[0] > tail_node[0]:
            tail_node[0] += 1
        else:
            tail_node[0] -= 1
        if head_node[1] > tail_node[1]:
            tail_node[1] += 1
        else:
            tail_node[1] -= 1
    elif abs(dist_vec[0]) > 1:
        if head_node[0] > tail_node[0]:
            tail_node[0] += 1
            tail_node[1] = head_node[1]
        else:
            tail_node[0] -= 1
            tail_node[1] = head_node[1]
    elif abs(dist_vec[1]) > 1:
        if head_node[1] > tail_node[1]:
            tail_node[1] += 1
            tail_node[0] = head_node[0]
        else:
            tail_node[1] -= 1
            tail_node[0] = head_node[0]
        
    return tail_node
```

```python
lst_visited = [(0, 0)]

with open(logfilename, "w") as log1:
    for instruction in instructions:
        direction = instruction[0]
        movement = instruction[1]
        for _ in range(movement):
            if direction == 'R':
                snake[0][0] += 1
            elif direction == 'L':
                snake[0][0] -= 1
            elif direction == 'U':
                snake[0][1] += 1
            elif direction == 'D':
                snake[0][1] -= 1

            for i in range(len(snake)-1):
                snake[i+1] = move_tail(snake[i], snake[i+1])
                log1.write(f"({direction}, {movement})[{i}] => {snake}\n")

            #print(f"Last visited: {lst_visited[-1]}, Current tail: {tuple(snake[-1])}")
            if lst_visited[-1] != tuple(snake[-1]):
                lst_visited.append(tuple(snake[-1]))
```

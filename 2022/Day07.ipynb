```python
dict_directory = {}
```

```python
dict_directory["/"] = {}
```

```python
dict_directory
```

```python
import re
```

```python
total_sz = 0
curr_dir = ""
for l in lst_input:
    ls = re.match(r"^\$ (ls)$", l)
    if ls:
        continue
    dr = re.match(r"^dir (.+)$", l)
    if dr:
        #print(f"Found subdirectory {dr.group(1)}")
        dict_directory[curr_dir + dr.group(1) + "/"] = {}
    sz = re.match(r"^([0-9]+) (.+)$", l)
    if sz:
        total_sz += int(sz.group(1))
        dict_directory[curr_dir][sz.group(2)] = sz.group(1)
    cd = re.match(r"^\$ cd (.+)$", l)
    if cd:
        if cd.group(1) == "/":
            curr_dir = "/"
        elif cd.group(1) == "..":
            curr_dir = "/".join(curr_dir.split("/")[:-2]) + "/"
        else:
            curr_dir = curr_dir + cd.group(1) + "/"
```

```python
dict_directory
```

```python
dict_sizes = {}
for x in dict_directory.items():
    dict_sizes[x[0]] = sum([int(f) for f in x[1].values()])
dict_sizes
```

```python
for s in dict_sizes.items():
    #print("=" * 20)
    dirs = s[0].split("/")[1:-1]
    #print(dirs)
    if len(dirs) > 1:
        #print(dirs)
        for i in range(1, len(dirs)):
            #print(i)
            #print("/" + "/".join(dirs[:i]) + "/")
            dict_sizes["/" + "/".join(dirs[:i]) + "/"] += s[1]
dict_sizes["/"] = total_sz
dict_sizes
```

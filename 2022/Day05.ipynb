```python
import re
import pandas as pd
```

```python
lst_input = str_input.split("\n")
```

```python
height = 0
data = {}
for l in lst_input:
    if len(l) == 0:
        break
    height += 1
    #l += " "
    if re.match(r".*\[.*", l) != None:
        l = re.sub(r"^    ", "[_] ", l)
        l = l.replace("    ", " [_]")
        l = re.sub(r"[\[\]]", "", l)
        l = re.sub(r"[\[\]]", "", l)
        l = re.sub(r" ", ".", l)
        print(l)
        data[height] = l.split(".")

print(height)

instructions = []
for l in lst_input[height + 1:]:
    l = [int(x) for x in (l.replace("move ", "").replace(" from ", ".").replace(" to ", ".")).split(".")]
    instructions.append(l)
    #print(l)
#print(instructions)

rows = height - 1
cols = int("".join(lst_input[height-1].split(" "))[-1:])
print(f'Columns = {cols}')
print(f'Rows = {rows}')

df_data = pd.DataFrame(data).T
df_data.columns = list(range(1, cols+1))
df_data
```

# Part 1

```python
dict_data = df_data.loc[::-1].to_dict('list')
dict_data
```

```python
for x in dict_data.items():
    dict_data[x[0]] = [y for y in x[1] if y != "_"]
dict_data
```

```python
for i in instructions:
    src = dict_data[i[1]][-i[0]:][::-1]
    trg = dict_data[i[2]]
    trg.extend(src)
    dict_data[i[2]] = trg
    dict_data[i[1]] = dict_data[i[1]][:-i[0]]
    #break

dict_data
```

```python
"".join([v[-1] for v in dict_data.values()])
```

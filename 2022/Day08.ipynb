```python
scenic_value = {}
#for t in zip(hidden_trees.keys(), hidden_trees.values()):
for t in zip(internal_trees.keys(), internal_trees.values()):
    xW = 0
    xE = 0
    yN = 0
    yS = 0

    r = t[0][0]

    counter = 0

    for c in range(t[0][1]):
        counter += 1
        #print(f"({t[0][0]}, {t[0][1] - c - 1}) => {df_trees.loc[r,t[0][1] - c - 1]}, counter = {counter}")
        if df_trees.loc[r,t[0][1] - c - 1] >= t[1]:
            break
    xW = counter

    counter = 0
    for c in range(99 - t[0][1] - 1):
        counter += 1
        #print(f"({t[0][0]}, {t[0][1] + c + 1}) => {df_trees.loc[r,t[0][1] + c + 1]}, counter = {counter}")
        if df_trees.loc[r,t[0][1] + c + 1] >= t[1]:
            break
    xE = counter

    c = t[0][1]

    counter = 0
    for r in range(t[0][0]):
        counter += 1
        #print(f"({t[0][0] - r - 1}, {t[0][1]}) => {df_trees.loc[t[0][0] - r - 1, c]}, counter = {counter}")
        if df_trees.loc[t[0][0] - r - 1, c] >= t[1]:
            break
    xN = counter

    counter = 0
    for r in range(99 - t[0][0] - 1):
        counter += 1
        #print(f"({t[0][0] - r + 1}, {t[0][1]}) => {df_trees.loc[t[0][0] + r + 1,c]}, counter = {counter}")
        if df_trees.loc[t[0][0] + r + 1,c] >= t[1]:
            break
    xS = counter

    scenic_value[t[0]] = [xW * xE * xN * xS, t[1], [xW, xE, xN, xS]]
```

```python
lst_input = str_input.split("\n")
instructions = [[y[0], int(y[1])] if len(y) > 1 else [y[0], ""] for y in [x.split(" ") for x in lst_input]]
instructions[:20]
```

# Part 1

```python
cycle_count = 0
reg_val = 1
reg_cycle_vals = {cycle_count: (reg_val, reg_val)}

with open(logfilename, "w") as log1:
    for instruction in instructions:
        log1.write(f"{'='*20}\n({instruction[0]}, {instruction[1]})\n{'-'*20}\n")
        if instruction[0] == 'noop':
            cycle_count += 1
            reg_cycle_vals[cycle_count] = (reg_cycle_vals[cycle_count-1][1], reg_cycle_vals[cycle_count-1][1])
            log1.write(f"{cycle_count} => {reg_cycle_vals[cycle_count]}\n")
        else:
            cycle_count += 1
            reg_cycle_vals[cycle_count] = (reg_cycle_vals[cycle_count-1][1], reg_cycle_vals[cycle_count-1][1])
            log1.write(f"{cycle_count} => {reg_cycle_vals[cycle_count]}\n")
            cycle_count += 1
            reg_cycle_vals[cycle_count] = (
                reg_cycle_vals[cycle_count-1][1], reg_cycle_vals[cycle_count-1][1] + instruction[1])
            log1.write(f"{cycle_count} => {reg_cycle_vals[cycle_count]}\n")

print(reg_cycle_vals)
```
